```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
unique_school_names = school_data_complete["school_name"].unique()
number_of_schools = len(unique_school_names)


# Calculate the total number of students
number_of_students = school_data_complete["student_name"].count()


# Calculate the total budget
total_budget = school_data["budget"].sum()


# Calculate the average maths score
average_maths_scores = round(school_data_complete["maths_score"].mean(),6)


average_reading_scores = round(school_data_complete["reading_score"].mean(),6)


# Calculate the percentage of students with a passing maths score (50 or greater)
passing_maths_students = school_data_complete.loc[school_data_complete["maths_score"] >= 50,:]
percentage_passing_maths_students = round(len(passing_maths_students)/len(school_data_complete["maths_score"]) * 100,6)


# Calculate the percentage of students with a passing reading score (50 or greater)
passing_reading_students = school_data_complete.loc[school_data_complete["reading_score"] >= 50,:]
percentage_passing_reading_students = round(len(passing_reading_students)/len(school_data_complete["reading_score"]) * 100,6)



# Calculate the percentage of students who passed maths and reading (% Overall Passing)
total_passing_maths_and_reading = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50),:]
percentage_passed_maths_and_reading = round(len(total_passing_maths_and_reading)/number_of_students * 100,6)


# Create a dataframe to hold the above results

local_govt_area_summary_df = pd.DataFrame({"Total Schools":[number_of_schools],
                            "Total Students":[number_of_students],
                            "Total Budget":[total_budget],
                            "Average Maths Score":[average_maths_scores],
                            "Average Reading Score":[average_reading_scores],
                            "% Passing Maths":[percentage_passing_maths_students],
                            "% Passing Reading":[percentage_passing_reading_students],
                            "% Overall Passing":[percentage_passed_maths_and_reading]})


# Optional: give the displayed data cleaner formatting
local_govt_area_summary_df["Total Students"] = local_govt_area_summary_df["Total Students"].map("{:,}".format)
local_govt_area_summary_df["Total Budget"] = local_govt_area_summary_df["Total Budget"].map("${:,}".format)
local_govt_area_summary_df["Average Maths Score"] = local_govt_area_summary_df["Average Maths Score"].map("{:.2f}".format)
local_govt_area_summary_df["Average Reading Score"] = local_govt_area_summary_df["Average Reading Score"].map("{:.2f}".format)
local_govt_area_summary_df["% Passing Maths"] = local_govt_area_summary_df["% Passing Maths"].map("{:.2f}".format)
local_govt_area_summary_df["% Passing Reading"] = local_govt_area_summary_df["% Passing Reading"].map("{:.2f}".format)
local_govt_area_summary_df["% Overall Passing"] = local_govt_area_summary_df["% Overall Passing"].map("{:.2f}".format)

local_govt_area_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Stored the school names by setting the index as school_name
school_names = school_data_complete.set_index("school_name").groupby(["school_name"])

# Stored the type according to school_name index
school_type = school_data.set_index("school_name")["type"]

# Total Students each school have
total_students_per_school = school_data_complete["school_name"].value_counts()

# Total School Budget
total_school_budget = school_data.set_index("school_name")["budget"].astype('float')

# Per Student Budget
per_student_budget = total_school_budget/total_students_per_school

# Average Maths Score
average_maths_score_per_school = school_names["maths_score"].mean()

# Average Reading Score
average_reading_score_per_school = school_names["reading_score"].mean()

# % Passing Maths
students_fulfil_maths_criteria = school_data_complete.loc[school_data_complete["maths_score"] >= 50,:]
passing_maths_per_school= students_fulfil_maths_criteria["school_name"].value_counts()
percentage_passing_maths_per_school = passing_maths_per_school/total_students_per_school*100

# % Passing Reading
students_fulfil_reading_criteria = school_data_complete.loc[school_data_complete["reading_score"] >= 50,:]
passing_reading_per_school= students_fulfil_reading_criteria["school_name"].value_counts()
percentage_passing_reading_per_school = passing_reading_per_school/total_students_per_school*100

# % Overall Passing (The percentage of students that passed maths and reading.)
students_fulfil_overall_criteria = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50),:]
passing_overall_per_school= students_fulfil_overall_criteria["school_name"].value_counts()
percentage_passing_overall_per_school = passing_overall_per_school/total_students_per_school*100

# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type":school_type,
                                    "Total Students":total_students_per_school,
                                  "Total School Budget":total_school_budget,
                                  "Per Student Budget":per_student_budget,
                                  "Average Maths Score":average_maths_score_per_school,
                                  "Average Reading Score":average_reading_score_per_school,
                                  "% Passing Maths":percentage_passing_maths_per_school,
                                  "% Passing Reading":percentage_passing_reading_per_school,
                                  "% Overall Passing":percentage_passing_overall_per_school
                                })


#Applying desired formatting
school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,}".format)
school_summary_df["Average Maths Score"] = school_summary_df["Average Maths Score"].map("{:.2f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:.2f}".format)
school_summary_df["% Passing Maths"] = school_summary_df["% Passing Maths"].map("{:.2f}".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:.2f}".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("{:.2f}".format)



school_summary_df.head(15)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Created bins as per the requirement
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

#assigning school_summary_df to new dataframe to perform the below operations
school_spending_df = school_summary_df

#assigning bins to the new column "Spending Ranges (Per Student)"
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Total School Budget"].str.strip("$").apply(lambda x: float(x.replace (',', '')))/school_spending_df["Total Students"].apply(lambda x: float(x.replace (',', ''))),spending_bins,labels=spending_labels)

#Change the datatype from object to float for the columns to perform mean()
school_spending_df = school_spending_df.astype({"Average Maths Score": float,"Average Reading Score":float, "% Passing Maths":float,"% Passing Reading":float,"% Passing Reading":float,"% Overall Passing":float }, errors='raise')

#Calculate average accordinbg to bins
average_spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
average_spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
average_spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
average_spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
average_overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

#Assigning values to the dataframe
spending_summary_df = pd.DataFrame({"Average Maths Score":average_spending_math_scores,
                                 "Average Reading Score":average_spending_reading_scores,
                                 "% Passing Maths":average_spending_passing_math,
                                 "% Passing Reading":average_spending_passing_reading,
                                 "% Overall Passing":average_overall_passing_spending})
#Applying desired formatting
spending_summary_df["Average Maths Score"] = spending_summary_df["Average Maths Score"].map("{:.2f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:.2f}".format)
spending_summary_df["% Passing Maths"] = spending_summary_df["% Passing Maths"].map("{:.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:.2f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:.2f}".format)

spending_summary_df.head()
```
